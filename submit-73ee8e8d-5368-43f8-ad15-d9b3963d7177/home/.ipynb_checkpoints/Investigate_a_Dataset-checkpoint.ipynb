### Research Question 2 Movies with highest profit margins 

```python
#. Finding the movies with highest  profitability_count
def max_value(column_name):
    #for highest profit
    max_value_forcolumnentered= tmdbmovie_data[column_name].idxmax()
    
    #.Printing the details of the movie with the lowest profit, please note the id to uniquely identify the row
    max_value_forcolumn_details=pd.DataFrame(tmdbmovie_data.loc[max_value_forcolumnentered])
    
    return max_value_forcolumn_details

    
#calling the function
max_value('profitability_count')
```

### Research Question 3 Movies with least profit margins 

```python
#.Finding the movies with lowest profitability_count

def min_value(column_name):
    #for lowest profit
    min_value_forcolumnentered= tmdbmovie_data[column_name].idxmin()
    
    #. Printing the details of the movie with the lowest profit, please note the id to uniquely identify the row
    min_value_forcolumn_details=pd.DataFrame(tmdbmovie_data.loc[min_value_forcolumnentered])
    
    return min_value_forcolumn_details
    
#calling the function
min_value('profitability_count')
```

### Research Question 13 Directors and the summation of the profit made by them on the movies they directed. Note the directors here are the directors who directed movies whose profit is greater than average profit.

### Research Question 17 Release_Year Vs Profitability_count.
Understanding what year brought the highest profit. Observing the fluctuating trend in the profit earned based on year of release, understanding what year worked the best for the movies.

```python
#. Release_Year Vs Profitability_Count
year_list = tmdbmovie_data.groupby('release_year')['profitability_count'].sum()

#figure size(width, height)
plt.figure(figsize=(12,6), dpi = 150)

#on x-axis
plt.xlabel('Release Year of Movies the dataset', fontsize = 12)
#on y-axis
plt.ylabel('Profits earned on Movies', fontsize = 12)
#title of the line plot
plt.title('Representing Total Profits earned by all movies Vs Year Released.')

#plotting the graph
plt.plot(year_list)

#displaying the line plot
plt.show()
```

### Research Question 20 Most frequent cast for movies

```python
### Finding the most frequent cast for movies
def data_count(column_passed):
    
    data_count = tmdbmovie_data[column_passed].str.cat(sep = '|')
    
    #giving pandas series and storing the values separately
    data_count = pd.Series(data_count.split('|'))
    
    #arranging in descending order
    count_value = data_count.value_counts(ascending = False)
    
    return count_value

#variable to store the retured value
count_value = data_count('cast')
#printing values
count_value
```

```python
count_value.idxmax()
```
